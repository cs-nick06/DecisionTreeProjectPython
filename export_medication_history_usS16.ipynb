```python
# Task 16
# Export medication history to a text file

import state

def export_history_for_user(user_email, file_path):
    if user_email is None:
        return False

    user_history = DOSE_HISTORY.get(user_email, [])

    try:
        with open(file_path, "w", encoding="utf-8") as f:
            f.write(f"Medication history for {user_email}\n")
            f.write("-" * 40 + "\n")

            if not user_history:
                f.write("No recorded medication doses.\n")
            else:
                for entry in user_history:
                    med_id = entry.get("medicine_id", "")
                    med_name = entry.get("medicine_name", "")
                    taken_at = entry.get("taken_at", "")
                    line = f"Medicine ID: {med_id} | Name: {med_name} | Taken at: {taken_at}\n"
                    f.write(line)

        print(f"Medication history exported to {file_path}")
        return True

    except Exception as e:
        print("Error writing history file:", e)
        return False
```
